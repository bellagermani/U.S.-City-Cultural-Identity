Creating percentage of aer emp

```python
# converting all values to numerics
EMP = emp.apply(pd.to_numeric, errors='coerce')
AER_EMP = aer_emp.apply(pd.to_numeric, errors='coerce')

# initializing new df
new_emp = pd.DataFrame()
new_emp["MSA"] = emp["MSA"] 

# Looping through dfs
for year in range(2010, 2016):
    # Defining column names for the current year in both dfs
    total_col = f'Tot_emp_{year}'
    aer_col = f'Aer_emp_{year}'
    share_col = f'Percent_Emp_{year}'
    # Dividing current year and adding result in new column
    new_emp[share_col] = AER_EMP[aer_col] / EMP[total_col]
    new_emp[share_col] = new_emp[share_col] * 100

# Replacing any missing values with .
new_emp.fillna('.', inplace = True)

# Renaming cols
new_emp = new_emp.rename(columns={'Percent_Emp_2010': 'Percent Emp 2010', 'Percent_Emp_2011': 'Percent Emp 2011', 'Percent_Emp_2012': 'Percent Emp 2012', 'Percent_Emp_2013': 'Percent Emp 2013', 'Percent_Emp_2014': 'Percent Emp 2014', 'Percent_Emp_2015': 'Percent Emp 2015'})

new_emp
```

```python
# Listing DFs
FBPs = [fbp10,fbp11,fbp12,fbp13,fbp14,fbp15,fbp16,fbp17,fbp18,fbp19,fbp20,fbp21,fbp22]
# Listing cols to keep
keep = ['Geographic Area Name','Total!!Estimate!!Foreign-born population','Estimate!!Total!!Foreign-born population'] # switches to 'Estimate!!Total!!' in 2017

# dropping unnecessary cols
for fbp in FBPs:    
    for col in fbp.columns:
        if col not in keep:
            fbp.drop(col, axis = 1, inplace = True)

# renaming cols
for fbp in FBPs:
    fbp.rename(columns = {'Geographic Area Name':'MSA'}, inplace = True)
    fbp.rename(columns = {'Estimate!!Total!!Foreign-born population':'Foreign-born Pop'}, inplace = True)
    fbp.rename(columns = {'Total!!Estimate!!Foreign-born population':'Foreign-born Pop'}, inplace = True)
```

```python
# removing 'Metro Area' str from MSA col
for fbp in FBPs:
    fbp['MSA'] = fbp['MSA'].str.replace('Metro Area', '')
    fbp['MSA'] = fbp['MSA'].str.strip() # removing whitespace

# Replacing NaNs with .
for fbp in FBPs:
    fbp['Foreign-born Pop'] = fbp['Foreign-born Pop'].fillna('.')

# Adding corresponding years to each col before merging
# dict of dfs with corresponding years
FBP_DICT = {
    2010:fbp10,
    2011:fbp11,
    2012:fbp12,
    2013:fbp13,
    2014:fbp14,
    2015:fbp15,
    2016:fbp16,
    2017:fbp17,
    2018:fbp18,
    2019:fbp19,
    2020:fbp20,
    2021:fbp21,
    2022:fbp22
}

# initializing
merged_fbp = None 

# looping through dict, adding years to cols, & merging
for year, fbp in FBP_DICT.items():
    fbp_renamed = fbp.rename(columns={'Foreign-born Pop': f'Foreign-born Pop {year}'})  # add corresponding year to each col
    if merged_fbp is None: 
        merged_fbp = fbp_renamed 
    else:
        merged_fbp = pd.merge(merged_fbp, fbp_renamed, on='MSA')

# deleting unnecessary cols
merged_fbp = merged_fbp.drop(columns = ['Foreign-born Pop 2016','Foreign-born Pop 2017','Foreign-born Pop 2018','Foreign-born Pop 2019','Foreign-born Pop 2020','Foreign-born Pop 2021','Foreign-born Pop 2022'])

# Renaming cols
merged_fbp = merged_fbp.rename(columns={'Foreign-born Pop 2010': 'Fborn Pop 2010', 'Foreign-born Pop 2011': 'Fborn Pop 2011', 'Foreign-born Pop 2012': 'Fborn Pop 2012', 'Foreign-born Pop 2013': 'Fborn Pop 2013', 'Foreign-born Pop 2014': 'Fborn Pop 2014', 'Foreign-born Pop 2015': 'Fborn Pop 2015'})

# Making alphabetical by MSA
merged_fbp = merged_fbp.sort_values(by='MSA', ascending = True)
merged_fbp
```

```python
gdp = pd.read_csv('/Users/isabellagermani/Desktop/Projects/CulturalIdentity/Data/gmpRGDP.csv', low_memory=False)

# Dropping cols
gdp = gdp.drop('GeoFIPS', axis = 1)
gdp = gdp.drop('ComponentName', axis = 1)
gdp = gdp.drop('ComponentId', axis = 1)
gdp = gdp.drop('Region', axis = 1)
gdp = gdp.drop('IndustryClassification', axis = 1)
gdp = gdp.drop('IndustryId', axis = 1)
gdp = gdp.drop(columns = ['2001','2002','2003','2004','2005','2006','2007','2008','2009','2016'])

# Dropping if not an MSA
gdp = gdp[gdp['GeoName'].str.contains('Metropolitan Statistical Area', na = False)]

# Losing Unnecessary str in GeoName col
gdp['GeoName'] = gdp['GeoName'].str.replace(' (Metropolitan Statistical Area)','')
gdp['GeoName'] = gdp['GeoName'].str.replace(' *','')

# Only including total GDP and arts&entertainment portion of GDP
gdp = gdp[gdp['Description'].str.contains('All industry total|Arts, entertainment, and recreation|Performing arts, spectator sports, museums, and related activities', na=False)]

# Replacing any missing values with .
gdp.fillna('.', inplace = True)
# Replacing (D) values with .
gdp = gdp.replace('(D)', '.')
gdp = gdp.replace('(L)', '.')
gdp = gdp.replace('(NA)', '.')
gdp = gdp.replace('(NM)', '.')

# Renaming GeoName
gdp.rename(columns = {'GeoName':'MSA'}, inplace = True)

# Making alphabetical by MSA
gdp = gdp.sort_values(by='MSA', ascending = True)
gdp.head(20)

# Copying w only certain columns
GDP = gdp[['MSA', 'Description', '2010','2011','2012','2013','2014','2015']].copy()
GDP.head(15)
```

```python
# 2010
# Using numeric values from 2010 col & disregarding .s
GDP['2010'] = pd.to_numeric(GDP['2010'], errors='coerce')

# Initializing output list
Perc_GDP_2010 = []

# Looping through 2010 column stepping by groups of 3
for i in range(0, len(GDP) - 2, 3):  # up to third-to-last row
    # Checking that none of the values in the current group arnt NaN
    if pd.notna(GDP['2010'].iloc[i]) and pd.notna(GDP['2010'].iloc[i + 1]) and pd.notna(GDP['2010'].iloc[i + 2]):
        # Adding the last two values (arts & perf) and dividing by the first (total GDP)
        sum = GDP['2010'].iloc[i + 1] + GDP['2010'].iloc[i + 2]
        result = (sum / GDP['2010'].iloc[i]) * 100  # calculating arts + performance % GDP
        Perc_GDP_2010.append(result)  # Appending result
    else: # if any value in the group of 3 is . just append . to list
        Perc_GDP_2010.append('.')  

print((Perc_GDP_2010))
print(len((Perc_GDP_2010)))
```

# PIVOING & MERGING EVERYTHING TOGETHER

```python
# Pivoting each df

# Defining function to pivot
def pivot_to_long_format(df, variable_prefix):
    # Extracting year cols based on variable prefix
    year_columns = [col for col in df.columns if col.startswith(variable_prefix)]
    
    # Pivoting to long
    df_long = pd.melt(
        df, 
        id_vars=['MSA'], 
        value_vars=year_columns, 
        var_name='Year', 
        value_name=variable_prefix
    )
    
    # Extracting year from column name
    df_long['Year'] = df_long['Year'].str.extract(r'(\d{4})').astype(int)
    
    return df_long
```

```python
# Loading & processing each dataframe

# new_emp
new_emp = pd.read_csv('/Users/isabellagermani/Desktop/Projects/CulturalIdentity/NEW_EMP.csv')
df_new_emp_long = pivot_to_long_format(new_emp, 'Percent Emp')
df_new_emp_long.to_csv('/Users/isabellagermani/Desktop/Projects/CulturalIdentity/NEW_EMP_long.csv', index=False)

# merged_fbp
merged_fbp = pd.read_csv('/Users/isabellagermani/Desktop/Projects/CulturalIdentity/MERGED_FBP.csv')
df_merged_fbp_long = pivot_to_long_format(merged_fbp, 'Fborn Pop')
df_merged_fbp_long.to_csv('/Users/isabellagermani/Desktop/Projects/CulturalIdentity/MERGED_FBP_long.csv', index=False)

# merged_ld
merged_ld = pd.read_csv('/Users/isabellagermani/Desktop/Projects/CulturalIdentity/MERGED_LD.csv')
df_merged_ld_long = pivot_to_long_format(merged_ld, 'Percent Speak Other Lang')
df_merged_ld_long.to_csv('/Users/isabellagermani/Desktop/Projects/CulturalIdentity/MERGED_LD_long.csv', index=False)

# Perc_GDP
Perc_GDP = pd.read_csv('/Users/isabellagermani/Desktop/Projects/CulturalIdentity/PERC_GDP.csv')
df_perc_gdp_long = pivot_to_long_format(Perc_GDP, 'Perc GDP')
df_perc_gdp_long.to_csv('/Users/isabellagermani/Desktop/Projects/CulturalIdentity/PERC_GDP_long.csv', index=False)

# pat
pat = pd.read_csv('/Users/isabellagermani/Desktop/Projects/CulturalIdentity/PAT.csv')
df_pat_long = pivot_to_long_format(pat, 'Pat')
df_pat_long.to_csv('/Users/isabellagermani/Desktop/Projects/CulturalIdentity/PAT_long.csv', index=False)

# pop
pop = pd.read_csv('/Users/isabellagermani/Desktop/Projects/CulturalIdentity/POP.csv')
df_pop_long = pivot_to_long_format(pop, 'Pop')
df_pop_long.to_csv('/Users/isabellagermani/Desktop/Projects/CulturalIdentity/POP_long.csv', index=False)
```

```python
# Merging

# Loading long-form data from CSV files above
new_emp_long = pd.read_csv("/Users/isabellagermani/Desktop/Projects/CulturalIdentity/NEW_EMP_long.csv")
merged_fbp_long = pd.read_csv("/Users/isabellagermani/Desktop/Projects/CulturalIdentity/MERGED_FBP_long.csv")
merged_ld_long = pd.read_csv("/Users/isabellagermani/Desktop/Projects/CulturalIdentity/MERGED_LD_long.csv")
perc_gdp_long = pd.read_csv("/Users/isabellagermani/Desktop//Projects/CulturalIdentity/PERC_GDP_long.csv")
pat_long = pd.read_csv("/Users/isabellagermani/Desktop/Projects/CulturalIdentity/PAT_long.csv")
pop_long = pd.read_csv("/Users/isabellagermani/Desktop/Projects/CulturalIdentity/POP_long.csv")

# Creating CulturalID df from PAT_long (MSA and Year columns)
CulturalID = pat_long[['MSA', 'Year']].drop_duplicates()

# Defining a function to extract MSA (matching up to two letters to the right of the comma for consistency - some of the MSA columns have weird state borders and will specify 2 states so only match up to the first)
def extract_msa_match(msa):
    # *Finding position of the comma and taking string up to the second letter after 
    comma_pos = msa.find(',')
    if comma_pos != -1:
        return msa[:comma_pos + 3]  # Include the comma, space, & 2 letters after
    return msa 

# Another function that merges on the extracted MSA match
def merge_on_msa(CulturalID_df, other_df, col_name):
    # Creating the MSA_match column (applying funct)
    CulturalID_df['MSA_match'] = CulturalID_df['MSA'].apply(extract_msa_match)
    other_df['MSA_match'] = other_df['MSA'].apply(extract_msa_match)
    
    # Merging on new 'MSA_match' col
    merged_df = pd.merge(CulturalID_df, other_df[['MSA_match', 'Year', col_name]], 
                         on=['MSA_match', 'Year'], how='left')
    
    # Dropping temporary 'MSA_match' column
    merged_df.drop(columns=['MSA_match'], inplace=True)
    return merged_df

# Merging each long dataset into the USCID DataFrame
USCID = merge_on_msa(CulturalID, new_emp_long, 'Percent Emp')
USCID = merge_on_msa(USCID, merged_fbp_long, 'Fborn Pop')
USCID = merge_on_msa(USCID, merged_ld_long, 'Percent Speak Other Lang')
USCID = merge_on_msa(USCID, perc_gdp_long, 'Perc GDP')
USCID = merge_on_msa(USCID, pat_long, 'Pat')
USCID = merge_on_msa(USCID, pop_long, 'Pop')

# Getting rid of any misc . or na values AGAIN
USCID.replace(['.', 'NA'], np.nan, inplace=True)

# Saving & exporting final merged df to a new CSV
USCID.to_csv("/Users/isabellagermani/Desktop/Projects/CulturalIdentity/USCID.csv", index=False)
```
